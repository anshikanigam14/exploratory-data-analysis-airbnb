Data Inspection includes the following tasks:

<ul>
    <li> Checking the shape of the dataframe: The dataset contains more than 100,000 rows and 26 columns. </li>
    <li> Describing column statistics.</li>
    <li> Checking for missing values: Columns such as "house_rules", "last_review", "reviews_per_month", "minimum_nights", etc. have the highest number of missing values.</li>

```python
def dataframe_inspect(df):
    # Check the columns and shape of the dataframe
    df_shape = df.shape
    print(pretty_print_start+'Shape of the Airbnb dataframe:'+pretty_print_end,df_shape,'\n')
    print(pretty_print_start+'Columns of Airbnb dataframe:\n'+pretty_print_end,df.columns,'\n')
    # describe column statistics
    df_description = df.describe()
    print(pretty_print_start+'Description of Airbnb column statistics:\n'+pretty_print_end,df_description,'\n')
    # Count the missing values in each column:
    print(pretty_print_start+'Missing values of airbnb dataframe:\n'+pretty_print_end,df.isna().sum().sort_values(ascending=False).head(10),'\n')
    # unique values of columns
    print(pretty_print_start+'Unique room types are:\n'+pretty_print_end,df['room type'].unique(),'\n')
    print(pretty_print_start+'Unique neighbourhood groups are:\n'+pretty_print_end,df['neighbourhood group'].unique(),'\n')
    print(pretty_print_start+'Unique Construction years are:\n'+pretty_print_end,df['Construction year'].unique(),'\n')
```

    <b>Data Cleaning, Transformation, Reduction, and Normalization</b>
</font><br>
<font size=3 align='left'><br>
These tasks include the following:

<ul>
    <li> Clean the column names: Convert column names to lowercase and replace spaces with underscores.</li>
    <li> Clean columns: Replace special characters, trim white spaces, change columns to date format, and transform the neighbourhood group column.</li>
    <li> Check for missing values: Identify columns with a high number of missing values, such as house_rules, last_review, reviews_per_month, minimum_nights, etc.</li>
    <li> Substitute missing values: Replace NaN (null) values with appropriate values.</li>
    <li> Remove unnecessary columns: Drop columns that will not be used in the project.</li>
</ul>
</font>

```python
def clean_dataframe(df):
    # Duplicate the dataset and clean column names
    airbnb_raw = df
    airbnb_raw.columns = airbnb_raw.columns.str.lower()
    airbnb_raw.columns = [col.replace(' ', '_') for col in airbnb_raw.columns]

    # clean columns, remove special characters if any
    airbnb_raw['price']= airbnb_raw['price'].replace(r'[^\w\s]|_', '', regex=True)
    airbnb_raw['service_fee']= airbnb_raw['service_fee'].replace(r'[^\w\s]|_', '', regex=True).astype(float)
    airbnb_raw['minimum_nights']= airbnb_raw['minimum_nights'].replace(r'[^\w\s]|_', '', regex=True)
    airbnb_raw['host_identity_verified'] = airbnb_raw['host_identity_verified'].str.strip()
    airbnb_raw['neighbourhood_group'] = np.where(airbnb_raw['neighbourhood_group'] == 'manhatan', 'Manhattan', airbnb_raw['neighbourhood_group'])
    airbnb_raw['neighbourhood_group'] = np.where(airbnb_raw['neighbourhood_group'] == 'brookln', 'Brooklyn', airbnb_raw['neighbourhood_group'])
    airbnb_raw['last_review']= pd.to_datetime(airbnb_raw['last_review'])
    
    # Substitute missing values
    airbnb_raw['reviews_per_month'] = airbnb_raw["reviews_per_month"].fillna(0)
    airbnb_raw['calculated_host_listings_count'] = airbnb_raw['calculated_host_listings_count'].fillna('0').astype(int)
    airbnb_raw['minimum_nights'] = airbnb_raw['minimum_nights'].fillna(0.0).abs().astype(float).astype(int)
    airbnb_raw['construction_year'] = airbnb_raw['construction_year'].fillna(0.0).abs().astype(float).astype(int)
    airbnb_raw['price']= airbnb_raw['price'].fillna('0.0').astype(float).astype(int)
    airbnb_raw['review_rate_number'] = airbnb_raw['review_rate_number'].fillna(0.0).abs().astype(float).astype(int)
    
    # Data reduction
    airbnb_raw = airbnb_raw.drop(['country','country_code','license','cancellation_policy','service_fee','number_of_reviews','reviews_per_month','availability_365','instant_bookable'],axis=1)
    return airbnb_raw
```

```python
airbnb_clean = clean_dataframe(airbnb_raw)
print(pretty_print_start+'Clean Dataframe:\n'+pretty_print_end,airbnb_clean.head(20),'\n')
```

```python
def total_listings_by_room_type(df):
    df_subset = df[['room_type', 'id', 'neighbourhood_group']]
    df_agg = df_subset.groupby(
        ['neighbourhood_group', 'room_type'],as_index=False).agg(total_ids=('id', 'nunique')).sort_values(by = 'total_ids', ascending = False)
    print(pretty_print_start+'Aggregated dataframe for stacked bar chart:\n\n'+pretty_print_end,df_agg)
    fig = px.bar(
        df_agg[['room_type', 'total_ids', 'neighbourhood_group']],
        x='neighbourhood_group',
        y='total_ids',
        color='room_type',
        width=900,
        height=400,
        title='<em>Total Listings by Room Types Across Neighbourhood Groups</em>',
        template='plotly_dark',
        barmode = 'stack',
        labels={'total_ids': '<b>Total Listings</b>',
                'neighbourhood_group': '<b>Neighbourhood Group</b>',
                'room_type': '<b>Room Type</b>'
                 },
        color_discrete_sequence=px.colors.sequential.Agsunset
    )
    fig.update_layout(
        xaxis_title='Neighbourhood Group',
        yaxis_title='Total Listings',
        showlegend=True,
        font=dict(family='Courier New, monospace', size=15, color='white'),
        yaxis_showticklabels=False,
        xaxis=dict(showgrid=False),
        yaxis=dict(showgrid=False),
        legend_title='<b>Room Type</b>'
    )
    fig.show()
```

```python
# Total Listings by Neighbourhood Group
def total_listings_by_neighbourhood_group(df):
    df_subset = df[['id', 'neighbourhood_group', 'price']]
    df_agg = df_subset.groupby('neighbourhood_group', as_index=False).agg(total_ids=('id', 'nunique'))
    print(pretty_print_start+'Aggregated dataframe for donut chart:\n'+pretty_print_end,df_agg)
    fig = px.pie(
        df_agg,
        values='total_ids',
        names='neighbourhood_group',
        hole=0.6,
        width=900,
        height=400,
        color_discrete_sequence=px.colors.sequential.Agsunset,
        labels={
            'total_ids': '<b>Total Listings</b>',
            'neighbourhood_group': '<b>Neighbourhood Group</b>'}
    )
    fig.update_layout(
        title='<em>Total Listings by Neighbourhood Group</em>',
        template='plotly_dark',
        font=dict(family='Courier New, monospace', size=15, color='white'),
        legend_title='<b>Neighbourhood Group</b>'
    )
    fig.update_traces(
        textinfo='percent+label',
        textposition='inside',
        textfont=dict(size=15, color='white')
    )
    fig.show()
```

```python
# Top 10 hosts with highest listings
def top_hosts_per_listings(df):
    df_subset = df[
        [
            'name',
            'calculated_host_listings_count',
            'host_id',
            'host_name',
            'host_identity_verified',
            'neighbourhood_group'
        ]
    ]
    df_subset = df_subset[df_subset['host_identity_verified'] == 'verified']
    df_agg = df_subset.groupby(['host_name', 'neighbourhood_group'], as_index=False).agg(total_ids=('calculated_host_listings_count', 'sum'))
    df_agg = df_agg.sort_values(by='total_ids', ascending=False).head(10)
    print(pretty_print_start+'Aggregated dataframe for bar chart:\n'+pretty_print_end,df_agg)
    fig = px.bar(
        df_agg,
        x='host_name',
        y='total_ids',
        template='plotly_dark',
        width=900,
        height=400,
        text='total_ids',
        orientation='v',
        color_discrete_sequence=px.colors.sequential.Agsunset
    )
    fig.update_layout(
        xaxis_title='Total Listings',
        yaxis_title='Host ID',
        title='<em>Top 10 Hosts by Listings</em>',
        font=dict(family='Courier New, monospace', size=15, color='white'),
        xaxis=dict(showgrid=False),
        yaxis=dict(showgrid=False)
    )
    fig.update_traces(
        textposition='inside',
        texttemplate='<b>%{text:}</b>',
        textfont=dict(size=15, color='white'),
        hovertext=df_agg[['total_ids', 'host_name', 'neighbourhood_group']],
        hovertemplate='<b>Total Listings</b>: %{hovertext[0]}<br>'
        + '<b>Host Name</b>: %{hovertext[1]}<br>'
        + '<b>Neighbourhood Group</b>: %{hovertext[2]}<br>'
    )
    fig.show()
```

```python
# Total listings constructed every 4 years
def total_listings_by_construction_year(df):
    df_subset = df[['construction_year', 'id', 'neighbourhood_group']]
    df_subset = df_subset[df_subset['construction_year'] != 0]
    # create a new column with year buckets
    df_subset['construction_year_bucket'] = pd.cut(
        df_subset['construction_year'],
        bins=range(2003, 2027, 4),
        labels=['2003-2006', '2007-2010', '2011-2014', '2015-2018', '2019-2022']
    )
    df_agg = df_subset.groupby(['neighbourhood_group', 'construction_year_bucket'], as_index=False).agg(total_ids=('id', 'nunique'))
    print(pretty_print_start+'Aggregated dataframe for Line chart:\n'+pretty_print_end,df_agg)
    fig = px.line(
        df_agg,
        x='construction_year_bucket',
        y='total_ids',
        width=1000,
        height=400,
        template='plotly_dark',
        color='neighbourhood_group',
        text='total_ids',
        color_discrete_sequence=px.colors.sequential.Agsunset,
        labels={
            'total_ids': '<b>Total Listings</b>',
            'neighbourhood_group': '<b>Neighbourhood Group</b>',
            'construction_year_bucket': '<b>Construction Year Range</b>'
            }
    )
    fig.update_layout(
        yaxis_title='Total Listings Constructed',
        xaxis_title='Construction Year',
        title='<em>Trend Analysis of Listings Constructed every 4 Years across Neighbourhoods</em>',
        template='plotly_dark',
        font=dict(family='Courier New, monospace', size=15, color='white'),
        legend_title='<b>Neighbourhood Group</b>',
        xaxis=dict(showgrid=False),
        yaxis=dict(showgrid=False)
    )
    fig.show()
```

```python
# Density mapbox of Price of Listings
def density_mapbox(df):
    df_subset = df[['id', 'host_id', 'name', 'host_name', 'neighbourhood_group', 'price','last_review','lat','long']]
    df_subset['last_reviewed_year'] = pd.to_datetime(df_subset['last_review']).dt.strftime('%Y')
    df_subset.dropna(inplace=True)
    df_subset = df_subset[(df_subset['last_reviewed_year'] <= '2023')]
    df_subset.sort_values(by='last_reviewed_year', ascending=True, inplace=True)
    print(pretty_print_start+'Dataframe for Density Plot:\n'+pretty_print_end,df_subset.head(10))
    fig = px.density_mapbox(
        df_subset,
        lat='lat',
        lon='long',
        z='price',
        radius=10,
        color_continuous_scale=px.colors.sequential.Agsunset_r,
        animation_frame='last_reviewed_year',
        labels={'price': '<b>Price($)</b>', 'last_reviewed_year':'<b>Last Reviewed Year</b>','lat':'<b>Latitude</b>', 'long':'<b>Longitude</b>', 'neighbourhood_group':'<b>Neighbourhood Group</b>' },
        mapbox_style='carto-darkmatter',
        zoom=10,
        center=dict(lat=df_subset['lat'].mean(), lon=df_subset['long'].mean()),
        hover_name='neighbourhood_group'
    )
    fig.update_layout(
        title='<em>Airbnb Price Density Map - Animated by Year</em>',
        template='plotly_dark',
        font=dict(family='Courier New, monospace', size=15, color='white'),
        legend_title='<b>Price($)</b>'
    )
    # fig.update_traces(
    #     hovertext=df_subset[['neighbourhood_group']],
        
    # )
    fig.show()
```

```python
df_subset = airbnb_clean[['minimum_nights', 'id','neighbourhood_group','price','room_type']]
min_price = df_subset['price'].min()
max_price = df_subset['price'].max()
price_bins = list(range(min_price, max_price + 250, 250))
df_subset['price_buckets'] = pd.cut(df_subset['price'], bins=price_bins)
df_subset['minimum_nights_flag'] = np.where(df_subset['minimum_nights'] > 7, '> a week', df_subset['minimum_nights'])

df_agg_1 = df_subset.groupby(['minimum_nights_flag'],as_index=False).agg(total_ids=('id', 'nunique'))
df_agg_2 = df_subset.groupby(['minimum_nights_flag','neighbourhood_group'],as_index=False).agg(total_ids=('id', 'nunique'))
df_agg_3 = df_subset.groupby(['minimum_nights_flag','price_buckets'],as_index=False).agg(total_ids=('id', 'nunique'))
df_agg_4 = df_subset.groupby(['minimum_nights_flag','room_type'],as_index=False).agg(total_ids=('id', 'nunique'))

print(pretty_print_start+'Aggregated dataframe for Subplot-1:\n'+pretty_print_end,df_agg_1)
print(pretty_print_start+'Aggregated dataframe for Subplot-2:\n'+pretty_print_end,df_agg_2)
print(pretty_print_start+'Aggregated dataframe for Subplot-3:\n'+pretty_print_end,df_agg_3)
print(pretty_print_start+'Aggregated dataframe for Subplot-4:\n'+pretty_print_end,df_agg_4)
```

```python
def stacked_bar_subplots(df_1,df_2,df_3,df_4):
    # Create subplots with 2 rows and 2 columns
    fig, axes = plt.subplots(2, 2, figsize=(12, 12))
    sns.set_palette("rocket_r")

    # Plot 1: Number of listings and minimum nights flag with cancellation policy
    ax_1 = sns.barplot(x='minimum_nights_flag', y='total_ids', data=df_1, ax=axes[0,0],dodge=False,errorbar=None)
    axes[0,0].set_xlabel('Minimum Nights Flag')
    axes[0,0].set_ylabel('Number of Listings')
    axes[0,0].set_title('Total Listings by Rental Duration',fontweight='bold')
    for i in ax_1.containers:
        ax_1.bar_label(i,)

    # Plot 2: Number of listings and minimum nights flag with neighbourhood_group
    sns.barplot(x='minimum_nights_flag', y='total_ids', hue='neighbourhood_group', data=df_2, ax=axes[1,1],dodge=False,errorbar=None)
    axes[1,1].set_xlabel('Minimum Nights Flag')
    axes[1,1].set_ylabel('Number of Listings')
    axes[1,1].set_title('Rental Duration Variation with Neighbourhood Group',fontweight='bold')


    # Plot 3: Number of listings and minimum nights flag with price_buckets
    sns.barplot(x='minimum_nights_flag', y='total_ids', hue='price_buckets', data=df_3, ax=axes[0,1],dodge=False,errorbar=None)
    axes[0,1].set_xlabel('Minimum Nights Flag')
    axes[0,1].set_ylabel('Number of Listings')
    axes[0,1].set_title('Rental Duration Variation with Price',fontweight='bold')

    # Plot 4: Number of listings and minimum nights flag with room_type
    sns.barplot(x='minimum_nights_flag', y='total_ids', hue='room_type', data=df_4, ax=axes[1,0],dodge=False,errorbar=None)
    axes[1,0].set_xlabel('Minimum Nights Flag')
    axes[1,0].set_ylabel('Number of Listings')
    axes[1,0].set_title('Rental Duration Variation with Property Type',fontweight='bold')

    # Add legend
    axes[0,1].legend(title='Price Buckets', loc='upper left',fontsize = 5,title_fontsize=6)
    axes[1,0].legend(title='Room Types', loc='upper left',fontsize = 5,title_fontsize=6)
    axes[1,1].legend(title='Neighbourhood Group', loc='upper left',fontsize = 5,title_fontsize=6)

    # Adjust spacing between subplots
    plt.subplots_adjust(wspace=0.2,hspace=0.4)

    # Show the plot
    plt.show()
```

```python
stacked_bar_subplots(df_agg_1, df_agg_2, df_agg_3, df_agg_4)
```

```python
def scatter_plot(df):
    df_subset = df[['id','review_rate_number','neighbourhood_group']]
    df_agg = df_subset.groupby(
        ['neighbourhood_group', 'review_rate_number'], as_index=False
    ).agg(total_ids=('id', 'nunique'))
    print(pretty_print_start+'Aggregated dataframe for Scatter Plot:\n'+pretty_print_end,df_agg)
    fig = px.scatter(df_agg,x='neighbourhood_group', y='review_rate_number', 
                    size = 'total_ids',color = 'review_rate_number',
                    title='<em>Neighborhood Group Rating Distribution</em>',
                labels={'total_ids': '<b>Total Listings</b>',
            'neighbourhood_group': '<b>Neighbourhood Group</b>',
            'review_rate_number': '<b>Review Rate Number</b>'
                },
            template='plotly_dark',
            color_discrete_sequence=px.colors.sequential.Agsunset_r
            )
    fig.update_layout(
        xaxis_title='Neighbourhood Group',
        yaxis_title='Review Rate Number',
        font=dict(family='Courier New, monospace', size=15, color='white'),
        yaxis_showticklabels=False,
        xaxis=dict(showgrid=False),
        yaxis=dict(showgrid=False),
        legend_title='<b>Review Rate Number</b>'
    )
    fig.show()
```
